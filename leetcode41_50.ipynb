```python
class Solution42:
    def trap(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        if len(height) < 3:
            return 0
        a_max = max(height)
        max_index = height.index(a_max)
        return self.l_search(height, max_index) + self.r_search(height, max_index)
    def l_search(self, height, right):
        if right < 2:
            return 0
        l_max = max(height[:right])
        l_index = height.index(l_max)
        v = (right - l_index - 1) * l_max
        for i in range(l_index+1, right):
            v -= height[i]
        return v + self.l_search(height, l_index)
    def r_search(self, height, left):
        if left > len(height) - 3:
            return 0
        r_max = max(height[left+1:])
        for i in range(len(height)-1, left, -1):
            if height[i] == r_max:
                r_index = i
                break
        v = (r_index - left - 1) * r_max
        for j in range(left+1, r_index):
            v -= height[j]
        return v + self.r_search(height, r_index)
```

```python
class Solution43:
    def multiply(self, num1, num2):
        """
        :type num1: str
        :type num2: str
        :rtype: str
        """
        if num1 == '0' or num2 == '0':
            return '0'
        l1 = len(num1)
        l2 = len(num2)
        if l2 > l1:
            return self.multiply(num2, num1)
        nums = []
        for i in range(l2):
            nums.append(self.strMul(num1, num2[i], len(num2)-i-1))
        result = ''
        for num in nums:
            result = self.strAdd(result, num)
        return result
    def strMul(self, num1, s2, i):
        up = 0
        out = '0' * i
        for j in range(len(num1)-1, -1, -1):
            num = int(num1[j]) * int(s2) + up
            out = str(num%10) + out
            up = num // 10
        if up > 0:
            out = str(up) + out
        return out
    def strAdd(self, s1, s2):
        out = ''
        up = 0
        l1 = len(s1)
        l2 = len(s2)
        if l1 == 0:
            return s2
        if l2 == 0:
            return s1
        i = l1 - 1
        j = l2 - 1
        while i >= 0 or j >= 0:
            if i < 0:
                num = int(s2[j]) + up
                j -= 1
            elif j < 0:
                num = int(s1[i]) + up
                i -= 1
            else:
                num = int(s1[i]) + int(s2[j]) + up
                i -= 1
                j -= 1
            out = str(num%10) + out
            up = num // 10
        if up == 1:
            out = '1' + out
        return out
```

```python
class Solution44:
    def isMatch(self, s, p):
        """
        :type s: str
        :type p: str
        :rtype: bool
        """
        m = len(s)
        n = len(p)
        dp = [[0 for _ in range(n+1)] for _ in range(m+1)]
        dp[0][0] = 1
        for k in range(1, n+1):
            if p[k-1] == '*':
                dp[0][k] = dp[0][k-1]
        for i in range(1, m+1):
            for j in range(1, n+1):
                if p[j-1] == '*':
                    dp[i][j] = max(dp[i][j-1], dp[i-1][j-1], dp[i-1][j])
                elif p[j-1] == s[i-1] or p[j-1] == '?':
                    dp[i][j] = dp[i-1][j-1]
        if dp[m][n] == 1:
            return True
        return False
```

```python
class Solution45:
    def jump(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        step = 0
        last_pos = 0
        max_pos = 0
        for i in range(len(nums)):
            if i > last_pos:
                last_pos = max_pos
                step += 1
            max_pos = max(max_pos, nums[i]+i)
        return step
```

```python
class Solution46:
    def permute(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        if nums == []:
            return []
        cnt = [0] * len(nums)
        self.out = []
        self.sel_next(nums, [], cnt)
        return self.out
    def sel_next(self, nums, n, cnt):
        if len(n) == len(nums):
            self.out.append(n[:])
            return
        for i in range(len(nums)):
            if cnt[i] == 0:
                n.append(nums[i])
                cnt[i] = 1
                self.sel_next(nums, n, cnt)
                cnt[i] = 0
                n.pop()
```

```python
class Solution47:
    def permuteUnique(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        if nums == []:
            return []
        cnt = [0] * len(nums)
        self.out = []
        self.sel_next(nums, [], cnt)
        return self.out
    def sel_next(self, nums, n, cnt):
        if len(n) == len(nums):
            if n not in self.out:
                self.out.append(n[:])
            return
        for i in range(len(nums)):
            if cnt[i] == 0:
                n.append(nums[i])
                cnt[i] = 1
                self.sel_next(nums, n, cnt)
                cnt[i] = 0
                n.pop()
```

```python
class Solution48:
    def rotate(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: void Do not return anything, modify matrix in-place instead.
        """
        for i in range(len(matrix)//2):
            matrix[i], matrix[len(matrix)-i-1] = matrix[len(matrix)-i-1], matrix[i]
        for j in range(len(matrix)-1):
            k = j + 1
            while k < len(matrix):
                matrix[j][k], matrix[k][j] = matrix[k][j], matrix[j][k]
                k += 1
```

```python
class Solution49:
    def groupAnagrams(self, strs):
        """
        :type strs: List[str]
        :rtype: List[List[str]]
        """
        res = {}
        for item in strs:
            k = ''.join(sorted(item))
            if k not in res:
                res[k] = []
            res[k].append(item)
        return [res[x] for x in res]
```
