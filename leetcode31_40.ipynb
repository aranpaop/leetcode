```python
class Solution31:
    def nextPermutation(self, nums):
        """
        :type nums: List[int]
        :rtype: void Do not return anything, modify nums in-place instead.
        """
        l = len(nums)
        if l > 1:
            i = l - 2
            while i >= 0:
                j = l - 1
                while j > i:
                    if nums[i] < nums[j]:
                        nums[i], nums[j] = nums[j], nums[i]
                        nums[i+1:] = sorted(nums[i+1:])
                        return
                    j -= 1
                i -= 1
            nums.reverse()
            return
```

```python
class Solution32:
    def longestValidParentheses(self, s):
        """
        :type s: str
        :rtype: int
        """
        a = len(s)
        if a < 2:
            return 0
        maxlen = 0
        dp = [0 for _ in range(a)]
        for i in range(1, a):
            pos = i - 1 -dp[i-1]
            if s[i]==')' and s[i-1] == '(' and i - 2 >= 0:
                dp[i] = dp[i-2] + 2
            elif s[i] == ')' and pos >= 0 and s[pos] == '(':
                dp[i] = dp[i-1] + 2
                if pos - 1 >= 0:
                    dp[i] += dp[i-dp[i-1]-2]    
        return max(dp)
```

```python
class Solution33:
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        if nums == []:
            return -1
        left = 0
        right = len(nums) - 1
        return self.br_search(nums, target, left, right)
    def br_search(self, nums, target, left, right):
        if left == right:
            if nums[left] == target:
                return left
            return -1
        if nums[left] < nums[right]:
            if target < nums[left] or target > nums[right]:
                return -1
            mid = (left + right) // 2
            if nums[mid] == target:
                return mid
            elif nums[mid] > target:
                return self.br_search(nums, target, left, mid-1)
            else:
                return self.br_search(nums, target, mid+1, right)
        if nums[left] > nums[right]:
            if target > nums[right] and target < nums[left]:
                return -1
            mid = (left + right) // 2
            if nums[mid] == target:
                return mid
            if nums[mid] >= nums[left]:
                if target < nums[mid] and target >= nums[left]:
                    return self.br_search(nums, target, left, mid-1)
                if target > nums[mid] or target <= nums[right]:
                    return self.br_search(nums, target, mid+1, right)
            if nums[mid] <= nums[right]:
                if target < nums[mid] or target >= nums[left]:
                    return self.br_search(nums, target, left, mid-1)
                if target > nums[mid] and target <= nums[right]:
                    return self.br_search(nums, target, mid+1, right)
```

```python
class Solution34:
    def searchRange(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        if nums == []:
            return [-1, -1]
        left = 0
        right = len(nums) - 1
        return [self.lo_search(nums, target, left, right), self.hi_search(nums, target, left, right)]
    def lo_search(self, nums, target, left, right):
        if left == right:
            if nums[left] == target:
                return left
            return -1
        mid = (left + right) // 2
        if nums[mid] == target:
            return self.lo_search(nums, target, left, mid)
        if nums[mid] < target:
            return self.lo_search(nums, target, mid+1, right)
        else:
            if mid == left:
                return -1
            return self.lo_search(nums, target, left, mid-1)
    def hi_search(self, nums, target, left, right):
        if left == right:
            if nums[left] == target:
                return left
            return -1
        mid = (left + right) // 2 + 1
        if nums[mid] == target:
            return self.hi_search(nums, target, mid, right)
        if nums[mid] > target:
            return self.hi_search(nums, target, left, mid-1)
        else:
            if mid == right:
                return -1
            return self.hi_search(nums, target, mid+1, right)
```

```python
class Solution35:
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        if nums == []:
            return 0
        if target <= nums[0]:
            return 0
        left = 0
        right = len(nums) - 1
        return self.binary_search(nums, target, left, right)
    def binary_search(self, nums, target, left, right):
        if right == left:
            if nums[left] >= target:
                return left 
            return left + 1
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        if nums[mid] > target:
            if mid == left:
                return left
            return self.binary_search(nums, target, left, mid-1)
        else:
            return self.binary_search(nums, target, mid+1, right)
```

```python
class Solution36:
    def isValidSudoku(self, board):
        """
        :type board: List[List[str]]
        :rtype: bool
        """
        for i in range(9):
            row = []
            for j in range(9):
                if board[i][j] == '.':
                    continue
                if board[i][j] in row:
                    return False
                row.append(board[i][j])
        for k in range(9):
            col = []
            for l in range(9):
                if board[l][k] == '.':
                    continue
                if board[l][k] in col:
                    return False
                col.append(board[l][k])
        for l in range(3):
            for m in range(3):
                squ = []
                for x in range(3*l, 3*l+3):
                    for y in range(3*m, 3*m+3):
                        if board[x][y] != '.':
                            if board[x][y] in squ:
                                return False
                            squ.append(board[x][y])
        return True
```

```python
class Solution37:
    def solveSudoku(self, board):
        """
        :type board: List[List[str]]
        :rtype: void Do not return anything, modify board in-place instead.
        """
        self.DFS(board)
        
    def isRowAndColValid(self, x, y, field, num):
        for i in range(9):
            if field[x][i] == str(num):
                return False
        for j in range(9):
            if field[j][y] == str(num):
                return False
        return True
 
    def isSquareValid(self, x, y, field, num):
        row, col = (x // 3) * 3, (y // 3) * 3
        for i in range(row, row + 3):
            for j in range(col, col + 3):
                if field[i][j] == str(num):
                    return False
        return True
 
    def getNext(self, field):
        for x in range(9):
            for y in range(9):
                if field[x][y] == ".":
                    return [x, y]
        return []
    
    def DFS(self, board):
            next = self.getNext(board)
            if not next:
                return True
            for i in range(1, 10):
                if self.isRowAndColValid(next[0], next[1], board, i) and self.isSquareValid(next[0], next[1], board, i):
                    board[next[0]][next[1]] = str(i)
                    if self.DFS(board):
                        return True
                    board[next[0]][next[1]] = "."
```

```python
class Solution38:
    def countAndSay(self, n):
        """
        :type n: int
        :rtype: str
        """
        s = '1'
        if n == 1:
            return s
        else:
            for _ in range(1, n):
                next_s = ''
                a = s[0]
                b = 0
                for i in range(len(s)):
                    if s[i] == a:
                        b += 1
                    else:
                        next_s += str(b)
                        next_s += a
                        a = s[i]
                        b = 1
                next_s += str(b)
                next_s += a
                s = next_s
        return s
```

```python
class Solution39:
    def combinationSum(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        candidates = list(set(candidates))
        candidates.sort()
        num_list = []
        out_list = []
        self.dfs(candidates, target, 0, num_list, out_list)
        return out_list
    def dfs(self, candidates, target, beg, num_list, out_list):
        if target == 0:
            out_list.append(num_list[:])
        elif target > 0:
            for i in range(beg, len(candidates)):
                num_list.append(candidates[i])
                self.dfs(candidates, target-candidates[i], i, num_list, out_list)
                num_list.pop()
        else:
            return
```

```python
class Solution40:
    def combinationSum2(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        candidates.sort()
        num_list = []
        out_list = []
        self.dfs(candidates, target, 0, num_list, out_list)
        return out_list
    def dfs(self, candidates, target, begin, num_list, out_list):
        if target == 0:
            if num_list not in out_list:
                out_list.append(num_list[:])
        elif target > 0:
            for i in range(begin, len(candidates)):
                num_list.append(candidates[i])
                self.dfs(candidates, target-candidates[i], i+1, num_list, out_list)
                num_list.pop()
        return
```
