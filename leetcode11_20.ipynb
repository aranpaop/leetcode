```python
class Solution11:
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        maxV = 0
        left = 0
        right = len(height) - 1
        while left < right:
            if height[left] <= height[right]:
                maxV = max(maxV, height[left]*(right-left))
                left += 1
            else:
                maxV = max(maxV, height[right]*(right-left))
                right -= 1
        return maxV
```

```python
class Solution12:
    def intToRoman(self, num):
        """
        :type num: int
        :rtype: str
        """
        out = ''
        a = num // 1000
        out += 'M' * a
        num = num % 1000
        a = num // 100
        if a < 4:
            out += 'C' * a
        elif a == 4:
            out += 'CD'
        elif a < 9:
            out = out + 'D' + 'C' * (a - 5)
        elif a == 9:
            out += 'CM'
        num = num % 100
        a = num // 10
        if a < 4:
            out += 'X' * a
        elif a == 4:
            out += 'XL'
        elif a < 9:
            out = out + 'L' + 'X' * (a - 5)
        elif a == 9:
            out += 'XC'
        a = num % 10
        if a < 4:
            out += 'I' * a
        elif a == 4:
            out += 'IV'
        elif a < 9:
            out = out + 'V' + 'I' * (a - 5)
        elif a == 9:
            out += 'IX'
        return out
```

```python
class Solution14:
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        if len(strs) == 0:
            return ''
        if len(strs) == 1:
            return strs[0]
        for i in range(len(strs)-1):
            length = min(len(strs[i]), len(strs[i+1]))
            if length == 0:
                return ''
            for j in range(length):
                if strs[i][j] != strs[i+1][j]:
                    strs[i+1] = strs[i+1][:j]
                    break
            if len(strs[i+1])>len(strs[i]):
                strs[i+1] = strs[i]
        return strs[-1]
```

```python
# exceed time limit
class Solution15:
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        a = []
        nums.sort()
        for i in range(len(nums)-2):
            left = i + 1
            right = len(nums) - 1
            while left < right:
                b = nums[i] + nums[left] + nums[right]
                if b == 0:
                    c = [nums[i], nums[left], nums[right]]
                    if c not in a:
                        a.append(c)
                    right -= 1
                    left += 1
                elif b > 0:
                    right -= 1
                else:
                    left += 1
        return a
```

```python
class Solution16:
    def threeSumClosest(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        sub = math.inf
        msub = -math.inf
        nums.sort()
        for i in range(len(nums)-2):
            left = i + 1
            right = len(nums) - 1
            while left < right:
                a = nums[i] + nums[left] + nums[right] - target
                if a == 0:
                    return target
                elif a > 0:
                    if a < sub:
                        sub = a
                    right -= 1
                else:
                    if a > msub:
                        msub = a
                    left += 1
        if -msub < sub:
            return target + msub
        else:
            return target + sub
```

```python
class Solution17:
    def letterCombinations(self, digits):
        """
        :type digits: str
        :rtype: List[str]
        """
        letters = [['a', 'b', 'c'], ['d', 'e', 'f'], ['g', 'h', 'i'], ['j', 'k', 'l'], ['m', 'n', 'o'], ['p', 'q', 'r', 's'], ['t', 'u', 'v'], ['w', 'x', 'y', 'z']]
        if len(digits) == 0:
            return []
        res = ['']
        for i in range(len(digits)):
            letter = letters[int(digits[i])-2]
            out = []
            for i in range(len(res)):
                for j in range(len(letter)):
                    out.append(res[i]+letter[j])
            res = out
        return out
```

```python
class Solution18:
    def fourSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        nums.sort()
        a = []
        for i in range(len(nums)-3):
            for j in range(i+1, len(nums)-2):
                left = j + 1
                right = len(nums) - 1
                while left < right:
                    b = nums[i] + nums[j] + nums[left] + nums[right] - target
                    if b == 0:
                        c = [nums[i], nums[j], nums[left], nums[right]]
                        if c not in a:
                            a.append(c)
                        left += 1
                        right -= 1
                    elif b > 0:
                        right -= 1
                    else:
                        left += 1
        return a
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution19:
    def removeNthFromEnd(self, head, n):
        """
        :type head: ListNode
        :type n: int
        :rtype: ListNode
        """
        p1 = ListNode(0)
        p1.next = head
        p2 = p1
        p3 = p1
        for _ in range(n):
            p1 = p1.next
        while p1.next != None:
            p1 = p1.next
            p2 = p2.next
        p2.next = p2.next.next
        return p3.next
```

```python
class Solution20:
    def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """
        if s == '':
            return True
        stack =[]
        for i in range(len(s)):
            if s[i] == '(' or s[i] == '[' or s[i] == '{':
                stack.append(s[i])
            else:
                if len(stack) == 0:
                    return False
                if s[i] == ')':
                    if stack[-1] == '(':
                        stack.pop()
                    else:
                        return False
                elif s[i] == ']':
                    if stack[-1] == '[':
                        stack.pop()
                    else:
                        return False
                elif s[i] == '}':
                    if stack[-1] == '{':
                        stack.pop()
                    else:
                        return False
        if len(stack) == 0:
            return True
        return False
```
