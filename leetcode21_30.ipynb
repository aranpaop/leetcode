```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution21:
    def mergeTwoLists(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        l3 = ListNode(0)
        p = l3
        while l1 != None or l2 != None:
            if l1 == None:
                p.next = l2
                break
            elif l2 == None:
                p.next = l1
                break
            else:
                if l1.val <= l2.val:
                    p.next = l1
                    p = p.next
                    l1 = l1.next
                else:
                    p.next = l2
                    p = p.next
                    l2 = l2.next
        return l3.next
```

```python
class Solution22:
    def generateParenthesis(self, n):
        """
        :type n: int
        :rtype: List[str]
        """
        res = []  
        self.generate(n, n, "", res)  
        return res  

    def generate(self, left, right, str, res):  
        if left == 0 and right == 0:  
            res.append(str)  
            return  
        if left > 0:  
            self.generate(left - 1, right, str + '(', res)  
        if right > left:  
            self.generate(left, right - 1, str + ')', res)  
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution23:
    def mergeKLists(self, lists):
        """
        :type lists: List[ListNode]
        :rtype: ListNode
        """
        nums = []
        for a in lists:
            while a != None:
                nums.append(a.val)
                a = a.next
        if len(nums) == 0:
            return None
        nums.sort()
        out = ListNode(nums[0])
        p = out
        for i in range(1, len(nums)):
            b = ListNode(nums[i])
            p.next = b
            p = p.next
        return out
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution24:
    def swapPairs(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if head == None:
            return None
        l = ListNode(0)
        l.next = head
        p = l
        while p.next != None:
            if p.next.next != None:
                nex = p.next.next.next
                res = p.next
                p.next = p.next.next
                p.next.next = res
                p.next.next.next = nex
                p = p.next.next
            else:
                break
        return l.next
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution25:
    def reverseKGroup(self, head, k):
        """
        :type head: ListNode
        :type k: int
        :rtype: ListNode
        """
        if k == 1:
            return head
        l = ListNode(0)
        l.next = head
        p1 = l
        p2 = l
        nex = l.next
        while nex != None:
            rev = 1
            for _ in range(k):
                if p1.next == None:
                    rev = 0
                    nex = None
                    break
                p1 = p1.next
            if rev == 1:
                bef = p2
                nex = p1.next
                p2 = p2.next
                p4 = p2
                p3 = p2.next
                p2.next = nex
                while p3 != p1:
                    res = p3.next
                    p3.next = p2
                    p2 = p3
                    p3 = res
                p3.next = p2
                bef.next = p3
                p2 = p4
                p1 = p4
        return l.next
```

```python
class Solution26:
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        i = 1
        while i < len(nums):
            if nums[i] == nums[i-1]:
                nums.pop(i)
            else:
                i += 1
        return len(nums)
```

```python
class Solution27:
    def removeElement(self, nums, val):
        """
        :type nums: List[int]
        :type val: int
        :rtype: int
        """
        i = 0
        while i < len(nums):
            if nums[i] == val:
                nums.pop(i)
            else:
                i += 1
        return len(nums)
```

```python
class Solution28:
    def strStr(self, haystack, needle):
        """
        :type haystack: str
        :type needle: str
        :rtype: int
        """
        if needle == '':
            return 0
        for i in range(len(haystack)-len(needle)+1):
            if haystack[i] == needle[0]:
                if len(needle) == 1:
                    return i
                match = 1
                for j in range(1, len(needle)):
                    if haystack[i+j] != needle[j]:
                        match = 0
                        break
                if match == 1:
                    return i
        return -1
```

```python
class Solution29:
    def divide(self, dividend, divisor):
        """
        :type dividend: int
        :type divisor: int
        :rtype: int
        """
        if dividend == - 2 ** 31 and divisor == -1:
            return 2 ** 31 - 1
        m1 = False
        if dividend < 0:
            m1 = True
            dividend = - dividend
        m2 = False
        if divisor < 0:
            m2 = True
            divisor = - divisor
        if dividend < divisor:
            return 0
        l1 = []
        l2 = []
        s = divisor
        i = 1
        while s <= dividend:
            l1.append(s)
            l2.append(i)
            s += s
            i += i
        s = l1[-1]
        i = l2[-1]
        j = len(l1) - 2
        while j >= 0:
            if s + l1[j] <= dividend:
                s += l1[j]
                i += l2[j]
            j -= 1
        if m1 ^ m2:
            return -i
        return i
```

```python
class Solution30:
    def findSubstring(self, s, words):
        """
        :type s: str
        :type words: List[str]
        :rtype: List[int]
        """
        l = len(s)
        if l == 0:
            return []
        n = len(words)
        if n == 0:
            return []
        w = len(words[0])
        o = n * w
        if l < o:
            return []
        out = []
        words.sort()
        for i in range(l-o+1):
            match = 1
            wi = []
            for j in range(n):
                wi.append(s[i+j*w:i+(j+1)*w])
            wi.sort()
            if wi == words:
                out.append(i)
        return out
```
